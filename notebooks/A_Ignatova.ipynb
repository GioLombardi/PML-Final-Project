```python
# 3. SDE specification
import torch
import torch.nn as nn
import numpy as np
from torchvision import datasets, transforms, utils
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

class GaussianFourierProjection(nn.Module):
    def __init__(self, embed_dim, scale=30.):
        super().__init__()
        self.W = nn.Parameter(torch.randn(embed_dim // 2) * scale, requires_grad=False)

    def forward(self, x):
        x_proj = x[:, None] * self.W[None, :] * 2 * np.pi
        return torch.cat([torch.sin(x_proj), torch.cos(x_proj)], dim=-1)

class Dense(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.dense = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        return self.dense(x)[..., None, None]

class ScoreNet(nn.Module):
    def __init__(self, marginal_prob_std, channels=[32, 64, 128, 256], embed_dim=256):
        super().__init__()
        self.embed = nn.Sequential(GaussianFourierProjection(embed_dim=embed_dim),
                                    nn.Linear(embed_dim, embed_dim))
        self.conv1 = nn.Conv2d(1, channels[0], 3, stride=1, bias=False)
        self.dense1 = Dense(embed_dim, channels[0])
        self.gnorm1 = nn.GroupNorm(4, num_channels=channels[0])
        self.conv2 = nn.Conv2d(channels[0], channels[1], 3, stride=2, bias=False)
        self.dense2 = Dense(embed_dim, channels[1])
        self.gnorm2 = nn.GroupNorm(32, num_channels=channels[1])
        self.conv3 = nn.Conv2d(channels[1], channels[2], 3, stride=2, bias=False)
        self.dense3 = Dense(embed_dim, channels[2])
        self.gnorm3 = nn.GroupNorm(32, num_channels=channels[2])
        self.conv4 = nn.Conv2d(channels[2], channels[3], 3, stride=2, bias=False)
        self.dense4 = Dense(embed_dim, channels[3])
        self.gnorm4 = nn.GroupNorm(32, num_channels=channels[3])
        self.tconv4 = nn.ConvTranspose2d(channels[3], channels[2], 3, stride=2, bias=False)
        self.dense5 = Dense(embed_dim, channels[2])
        self.tgnorm4 = nn.GroupNorm(32, num_channels=channels[2])
        self.tconv3 = nn.ConvTranspose2d(channels[2] + channels[2], channels[1], 3, stride=2, bias=False, output_padding=1)
        self.dense6 = Dense(embed_dim, channels[1])
        self.tgnorm3 = nn.GroupNorm(32, num_channels=channels[1])
        self.tconv2 = nn.ConvTranspose2d(channels[1] + channels[1], channels[0], 3, stride=2, bias=False, output_padding=1)
        self.dense7 = Dense(embed_dim, channels[0])
        self.tgnorm2 = nn.GroupNorm(32, num_channels=channels[0])
        self.tconv1 = nn.ConvTranspose2d(channels[0] + channels[0], 1, 3, stride=1)
        self.act = lambda x: x * torch.sigmoid(x)
        self.marginal_prob_std = marginal_prob_std

    def forward(self, x, t):
        embed = self.act(self.embed(t))
        h1 = self.conv1(x) + self.dense1(embed)
        h1 = self.act(self.gnorm1(h1))
        h2 = self.conv2(h1) + self.dense2(embed)
        h2 = self.act(self.gnorm2(h2))
        h3 = self.conv3(h2) + self.dense3(embed)
        h3 = self.act(self.gnorm3(h3))
        h4 = self.conv4(h3) + self.dense4(embed)
        h4 = self.act(self.gnorm4(h4))
        h = self.tconv4(h4) + self.dense5(embed)
        h = self.act(self.tgnorm4(h))
        h = self.tconv3(torch.cat([h, h3], dim=1)) + self.dense6(embed)
        h = self.act(self.tgnorm3(h))
        h = self.tconv2(torch.cat([h, h2], dim=1)) + self.dense7(embed)
        h = self.act(self.tgnorm2(h))
        h = self.tconv1(torch.cat([h, h1], dim=1))
        return h / self.marginal_prob_std(t)[:, None, None, None]

def marginal_prob_std(t, sigma=25.):
    return torch.sqrt((sigma**(2 * t) - 1.) / (2. * np.log(sigma)))

def diffusion_coeff(t, sigma=25.):
    return sigma**t

def Euler_Maruyama_sampler(score_model, marginal_prob_std, diffusion_coeff, batch_size=256, num_steps=1000, device='cuda', eps=1e-3):
    t = torch.ones(batch_size, device=device)
    init_x = torch.randn(batch_size, 1, 28, 28, device=device) * marginal_prob_std(t)[:, None, None, None]
    time_steps = torch.linspace(1., eps, num_steps, device=device)
    step_size = time_steps[0] - time_steps[1]
    x = init_x
    with torch.no_grad():
        for time_step in tqdm(time_steps):
            batch_time_step = torch.ones(batch_size, device=device) * time_step
            g = diffusion_coeff(batch_time_step)
            mean_x = x + (g**2)[:, None, None, None] * score_model(x, batch_time_step) * step_size
            x = mean_x + torch.sqrt(step_size) * g[:, None, None, None] * torch.randn_like(x)
    return x

# Training the model
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: (x - 0.5) * 2.0)
])

dataloader_train = torch.utils.data.DataLoader(
    datasets.MNIST('./mnist_data', download=True, train=True, transform=transform),
    batch_size=256,
    shuffle=True
)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = ScoreNet(marginal_prob_std).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

def train_sde(model, dataloader, optimizer, num_epochs=100):
    for epoch in range(num_epochs):
        for x, _ in tqdm(dataloader):
            x = x.to(device)
            optimizer.zero_grad()
            t = torch.rand(x.size(0), device=device)
            noise = torch.randn_like(x)
            perturbed_x = x + marginal_prob_std(t)[:, None, None, None] * noise
            score = model(perturbed_x, t)
            loss = torch.mean((score * marginal_prob_std(t)[:, None, None, None] + noise)**2)
            loss.backward()
            optimizer.step()
        print(f"Epoch {epoch + 1}, Loss: {loss.item():.4f}")

train_sde(model, dataloader_train, optimizer, num_epochs=50)

# Sampling
samples = Euler_Maruyama_sampler(model, marginal_prob_std, diffusion_coeff, batch_size=256, num_steps=1000, device=device)
samples = (samples + 1) / 2
samples = samples.clamp(0.0, 1.0)

# Visualize
samples_grid = utils.make_grid(samples[:36], nrow=6)
plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(samples_grid.permute(1, 2, 0).cpu().numpy(), cmap='gray')
plt.show()
```

```python
# 3b Experimenting with Continuous Diffusion Model
import torch
import torch.nn as nn
import numpy as np
from torchvision import datasets, transforms, utils
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define Gaussian Fourier Projection
class GaussianFourierProjection(nn.Module):
    def __init__(self, embed_dim, scale=30.):
        super().__init__()
        self.W = nn.Parameter(torch.randn(embed_dim // 2) * scale, requires_grad=False)

    def forward(self, x):
        x_proj = x[:, None] * self.W[None, :] * 2 * np.pi
        return torch.cat([torch.sin(x_proj), torch.cos(x_proj)], dim=-1)

# Dense Layer
class Dense(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.dense = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        return self.dense(x)[..., None, None]

# ScoreNet Model
class ScoreNet(nn.Module):
    def __init__(self, marginal_prob_std, channels=[32, 64, 128, 256], embed_dim=256):
        super().__init__()
        self.embed = nn.Sequential(GaussianFourierProjection(embed_dim=embed_dim),
                                    nn.Linear(embed_dim, embed_dim))
        self.conv1 = nn.Conv2d(1, channels[0], 3, stride=1, bias=False)
        self.dense1 = Dense(embed_dim, channels[0])
        self.gnorm1 = nn.GroupNorm(4, num_channels=channels[0])
        self.conv2 = nn.Conv2d(channels[0], channels[1], 3, stride=2, bias=False)
        self.dense2 = Dense(embed_dim, channels[1])
        self.gnorm2 = nn.GroupNorm(32, num_channels=channels[1])
        self.conv3 = nn.Conv2d(channels[1], channels[2], 3, stride=2, bias=False)
        self.dense3 = Dense(embed_dim, channels[2])
        self.gnorm3 = nn.GroupNorm(32, num_channels=channels[2])
        self.conv4 = nn.Conv2d(channels[2], channels[3], 3, stride=2, bias=False)
        self.dense4 = Dense(embed_dim, channels[3])
        self.gnorm4 = nn.GroupNorm(32, num_channels=channels[3])
        self.tconv4 = nn.ConvTranspose2d(channels[3], channels[2], 3, stride=2, bias=False)
        self.dense5 = Dense(embed_dim, channels[2])
        self.tgnorm4 = nn.GroupNorm(32, num_channels=channels[2])
        self.tconv3 = nn.ConvTranspose2d(channels[2] + channels[2], channels[1], 3, stride=2, bias=False, output_padding=1)
        self.dense6 = Dense(embed_dim, channels[1])
        self.tgnorm3 = nn.GroupNorm(32, num_channels=channels[1])
        self.tconv2 = nn.ConvTranspose2d(channels[1] + channels[1], channels[0], 3, stride=2, bias=False, output_padding=1)
        self.dense7 = Dense(embed_dim, channels[0])
        self.tgnorm2 = nn.GroupNorm(32, num_channels=channels[0])
        self.tconv1 = nn.ConvTranspose2d(channels[0] + channels[0], 1, 3, stride=1)
        self.act = lambda x: x * torch.sigmoid(x)
        self.marginal_prob_std = marginal_prob_std

    def forward(self, x, t):
        embed = self.act(self.embed(t))
        h1 = self.conv1(x) + self.dense1(embed)
        h1 = self.act(self.gnorm1(h1))
        h2 = self.conv2(h1) + self.dense2(embed)
        h2 = self.act(self.gnorm2(h2))
        h3 = self.conv3(h2) + self.dense3(embed)
        h3 = self.act(self.gnorm3(h3))
        h4 = self.conv4(h3) + self.dense4(embed)
        h4 = self.act(self.gnorm4(h4))
        h = self.tconv4(h4) + self.dense5(embed)
        h = self.act(self.tgnorm4(h))
        h = self.tconv3(torch.cat([h, h3], dim=1)) + self.dense6(embed)
        h = self.act(self.tgnorm3(h))
        h = self.tconv2(torch.cat([h, h2], dim=1)) + self.dense7(embed)
        h = self.act(self.tgnorm2(h))
        h = self.tconv1(torch.cat([h, h1], dim=1))
        return h / self.marginal_prob_std(t)[:, None, None, None]

# Variance Exploding SDE
sigma = 25.0

def ve_marginal_prob_std(t, sigma=sigma):
    return torch.sqrt((sigma**(2 * t) - 1.) / (2. * np.log(sigma)))

def ve_diffusion_coeff(t, sigma=sigma):
    return sigma**t

# Variance Preserving SDE
beta_min, beta_max = 0.1, 20.0

def vp_marginal_prob_std(t, beta_min=beta_min, beta_max=beta_max):
    beta_t = beta_min + t * (beta_max - beta_min)
    return torch.sqrt(1. - torch.exp(-2 * beta_t))

def vp_diffusion_coeff(t, beta_min=beta_min, beta_max=beta_max):
    beta_t = beta_min + t * (beta_max - beta_min)
    return torch.sqrt(2. * beta_t)

# Euler-Maruyama Sampler
def Euler_Maruyama_sampler(score_model, marginal_prob_std, diffusion_coeff, batch_size=256, num_steps=1000, device='cuda', eps=1e-3):
    t = torch.ones(batch_size, device=device)
    init_x = torch.randn(batch_size, 1, 28, 28, device=device) * marginal_prob_std(t)[:, None, None, None]
    time_steps = torch.linspace(1., eps, num_steps, device=device)
    step_size = time_steps[0] - time_steps[1]
    x = init_x
    with torch.no_grad():
        for time_step in tqdm(time_steps):
            batch_time_step = torch.ones(batch_size, device=device) * time_step
            g = diffusion_coeff(batch_time_step)
            mean_x = x + (g**2)[:, None, None, None] * score_model(x, batch_time_step) * step_size
            x = mean_x + torch.sqrt(step_size) * g[:, None, None, None] * torch.randn_like(x)
    return x

# ODE Sampler
def ode_reverse(score_model, marginal_prob_std, batch_size=256, num_steps=1000, device='cuda'):
    t = torch.ones(batch_size, device=device)
    init_x = torch.randn(batch_size, 1, 28, 28, device=device) * marginal_prob_std(t)[:, None, None, None]

    def ode_func(t, x):
        t_tensor = torch.tensor([t], device=device)
        x_tensor = torch.tensor(x.reshape(init_x.shape), device=device)
        score = score_model(x_tensor, t_tensor)
        g = ve_diffusion_coeff(t_tensor)
        return (score / g).cpu().numpy().ravel()

    x = init_x.cpu().numpy().ravel()
    time_steps = np.linspace(1., 1e-3, num_steps)
    solution = solve_ivp(ode_func, (1., 1e-3), x, t_eval=time_steps, method='RK45')
    samples = solution.y[:, -1].reshape(init_x.shape)
    return torch.tensor(samples, device=device)

# Training the Model
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: (x - 0.5) * 2.0)
])

dataloader_train = torch.utils.data.DataLoader(
    datasets.MNIST('./mnist_data', download=True, train=True, transform=transform),
    batch_size=256,
    shuffle=True
)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = ScoreNet(ve_marginal_prob_std).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

def train_sde(model, dataloader, optimizer, num_epochs=100):
    for epoch in range(num_epochs):
        for x, _ in tqdm(dataloader):
            x = x.to(device)
            optimizer.zero_grad()
            t = torch.rand(x.size(0), device=device)
            noise = torch.randn_like(x)
            perturbed_x = x + ve_marginal_prob_std(t)[:, None, None, None] * noise
            score = model(perturbed_x, t)
            loss = torch.mean((score * ve_marginal_prob_std(t)[:, None, None, None] + noise)**2)
            loss.backward()
            optimizer.step()
        print(f"Epoch {epoch + 1}, Loss: {loss.item():.4f}")

train_sde(model, dataloader_train, optimizer, num_epochs=100)

# Sampling with Euler-Maruyama
samples = Euler_Maruyama_sampler(model, ve_marginal_prob_std, ve_diffusion_coeff, batch_size=256, num_steps=1000, device=device)
samples = (samples + 1) / 2
samples = samples.clamp(0.0, 1.0)

# Visualize Samples
samples_grid = utils.make_grid(samples[:36], nrow=6)
plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(samples_grid.permute(1, 2, 0).cpu().numpy(), cmap='gray')
plt.show()
```

```python
# 4 “Classifier-free guidance
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torchvision import datasets, transforms, utils
from tqdm.auto import tqdm

# Define the Unet model (updated for Classifier-Free Guidance)
class ConditionalUnet(nn.Module):
    def __init__(self):
        super().__init__()
        nch = 2
        chs = [32, 64, 128, 256, 256]
        self._convs = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(2 + 10, chs[0], kernel_size=3, padding=1),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.MaxPool2d(kernel_size=2, stride=2),
                nn.Conv2d(chs[0], chs[1], kernel_size=3, padding=1),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.MaxPool2d(kernel_size=2, stride=2),
                nn.Conv2d(chs[1], chs[2], kernel_size=3, padding=1),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.MaxPool2d(kernel_size=2, stride=2, padding=1),
                nn.Conv2d(chs[2], chs[3], kernel_size=3, padding=1),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.MaxPool2d(kernel_size=2, stride=2),
                nn.Conv2d(chs[3], chs[4], kernel_size=3, padding=1),
                nn.ReLU(),
            ),
        ])
        self._tconvs = nn.ModuleList([
            nn.Sequential(
                nn.ConvTranspose2d(chs[4], chs[3], kernel_size=3, stride=2, padding=1, output_padding=1),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.ConvTranspose2d(chs[3] * 2, chs[2], kernel_size=3, stride=2, padding=1, output_padding=0),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.ConvTranspose2d(chs[2] * 2, chs[1], kernel_size=3, stride=2, padding=1, output_padding=1),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.ConvTranspose2d(chs[1] * 2, chs[0], kernel_size=3, stride=2, padding=1, output_padding=1),
                nn.ReLU(),
            ),
            nn.Sequential(
                nn.Conv2d(chs[0] * 2, chs[0], kernel_size=3, padding=1),
                nn.ReLU(),
                nn.Conv2d(chs[0], 1, kernel_size=3, padding=1),
            ),
        ])

    def forward(self, x: torch.Tensor, t: torch.Tensor, y: torch.Tensor) -> torch.Tensor:
        x2 = torch.reshape(x, (*x.shape[:-1], 1, 28, 28))
        tt = t[..., None, None].expand(*t.shape[:-1], 1, 28, 28)
        yy = y.view(-1, 10, 1, 1).expand(-1, -1, 28, 28)
        x2t = torch.cat((x2, tt, yy), dim=-3)
        signal = x2t
        signals = []
        for i, conv in enumerate(self._convs):
            signal = conv(signal)
            if i < len(self._convs) - 1:
                signals.append(signal)

        for i, tconv in enumerate(self._tconvs):
            if i == 0:
                signal = tconv(signal)
            else:
                signal = torch.cat((signal, signals[-i]), dim=-3)
                signal = tconv(signal)
        signal = torch.reshape(signal, (*signal.shape[:-3], -1))
        return signal


# Define the Classifier-Free Guided Diffusion Model
class ClassifierFreeGuidedDDPM(nn.Module):
    def __init__(self, network, T=100, beta_1=1e-4, beta_T=2e-2, guidance_strength=1.0):
        super().__init__()
        self.network = network
        self.T = T
        self.guidance_strength = guidance_strength
        self.register_buffer("beta", torch.linspace(beta_1, beta_T, T + 1))
        self.register_buffer("alpha", 1 - self.beta)
        self.register_buffer("alpha_bar", self.alpha.cumprod(dim=0))

    def forward_diffusion(self, x0, t, epsilon):
        alpha_bar_t = self.alpha_bar[t - 1]
        mean = torch.sqrt(alpha_bar_t) * x0
        std = torch.sqrt(1 - alpha_bar_t)
        return mean + std * epsilon

    @torch.no_grad()
    def sample(self, shape, y):
        xT = torch.randn(shape).to(self.beta.device)
        xt = xT
        for t in range(self.T, 0, -1):
            noise = torch.randn_like(xT) if t > 1 else 0
            t_tensor = torch.tensor(t).expand(xt.shape[0], 1).to(self.beta.device)

            eps_cond = self.network(xt, t_tensor, y)
            eps_uncond = self.network(xt, t_tensor, torch.zeros_like(y))

            eps_theta = eps_uncond + self.guidance_strength * (eps_cond - eps_uncond)

            alpha_t = self.alpha[t - 1]
            alpha_bar_t = self.alpha_bar[t - 1]
            beta_t = self.beta[t - 1]
            mean = 1 / torch.sqrt(alpha_t) * (xt - (1 - alpha_t) / torch.sqrt(1 - alpha_bar_t) * eps_theta)
            std = torch.sqrt(beta_t)
            xt = mean + std * noise

        return xt


# Prepare MNIST dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: (x - 0.5) * 2.0),
    transforms.Lambda(lambda x: x.flatten())
])

dataloader_train = torch.utils.data.DataLoader(
    datasets.MNIST('./mnist_data', download=True, train=True, transform=transform),
    batch_size=256, shuffle=True
)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
mnist_cond_unet = ConditionalUnet().to(device)
cond_model = ClassifierFreeGuidedDDPM(mnist_cond_unet, T=200, guidance_strength=1.5).to(device)


optimizer = torch.optim.Adam(cond_model.parameters(), lr=1e-3)
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, 0.9999)

def train_conditional_model(model, optimizer, scheduler, dataloader, epochs, device):
    progress_bar = tqdm(range(len(dataloader) * epochs), desc="Training")
    for epoch in range(epochs):
        model.train()
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)

            # One-hot encode labels
            y_onehot = torch.zeros(x.size(0), 10).to(device)
            y_onehot.scatter_(1, y.view(-1, 1), 1)

            # Sample random timesteps
            t = torch.randint(1, model.T + 1, (x.size(0), 1)).to(device)

            # Sample noise
            epsilon = torch.randn_like(x).to(device)

            # Compute forward diffusion
            xt = model.forward_diffusion(x, t, epsilon)

            # Compute the loss (mean squared error between epsilon and the model's prediction)
            optimizer.zero_grad()
            loss = nn.MSELoss()(epsilon, model.network(xt, t, y_onehot))
            loss.backward()
            optimizer.step()
            scheduler.step()

            progress_bar.update()
            progress_bar.set_postfix(epoch=epoch + 1, loss=loss.item())
        print(f"Epoch {epoch + 1}/{epochs} completed.")


train_conditional_model(cond_model, optimizer, scheduler, dataloader_train, epochs=100, device=device)

# Generate samples for digit 4
y_target = torch.zeros((10, 10)).to(device)
y_target[:, 4] = 1  # One-hot encode digit 4
samples = cond_model.sample((10, 28 * 28), y_target).cpu()
samples = (samples + 1) / 2

# Display the samples
grid = utils.make_grid(samples.reshape(-1, 1, 28, 28), nrow=10)
plt.title("Classifier-Free Guided Samples of Digit 4")
plt.imshow(transforms.functional.to_pil_image(grid), cmap="gray")
plt.show()
```
